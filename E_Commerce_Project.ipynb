```python
# Link to the dataset: https://www.kaggle.com/datasets/thedevastator/unlock-profits-with-e-commerce-sales-data?resource=download
import os
import zipfile

zip_file_path = r'D:\Learning\Project\E-Commerce Project\Amazon Sale Report.csv.zip'
extract_file_path = r'D:\Learning\Project\E-Commerce Project'
extracted_file = os.path.join(extract_file_path, 'Amazon Sale Report.csv')

# Check if the extracted file already exists
if os.path.exists(extracted_file):
    print("File already present.")
else:
    with zipfile.ZipFile(zip_file_path, 'r') as un_zip:
        un_zip.extractall(extract_file_path)
        print("Zip file extracted successfully.")
```

```python
# Load the CSV file into python
import pandas as pd

file_path = r'D:\Learning\Project\E-Commerce Project\Amazon Sale Report.csv'
df = pd.read_csv(file_path)

# Display some information about the dataset
df.info()
```

```python
# Check if the head of the data
df.head()
```

```python
# Now this data should be cleaned and one of the first step is to remove all the irrelevant columns
df_cleaned = df.drop(columns = ['index', 'Unnamed: 22'])
```

```python
df_cleaned.info()
```

```python
# Checking and removing any duplicates found in the data
duplicate_data = df_cleaned[df_cleaned.duplicated()]
print(duplicate_data)
duplicate_sum = df_cleaned.duplicated().sum()
print('Total duplicate values are: {}'.format(duplicate_sum))
```

```python
# Now these duplicate values need to be removed
df_cleaned_dups = df_cleaned.drop_duplicates()
```

```python
# Lets see if there are any duplicated values in the dataset
print('Checking for duplicate values after removing them: {} Found'.format(df_cleaned_dups.duplicated().sum()))
```

```python
# Check for missing values in each column
missing_values = df_cleaned_dups.isnull().sum()

# Display columns with their corresponding count of missing values
print(missing_values)
```

```python
# There are quite a few null values in the columns now lets replace the values or remove those values if the row can't be used with that missing value like amount or currency
df_cleaned_rm_rows = df_cleaned_dups.dropna(subset=['Amount', 'currency', 'ship-city', 'ship-state', 'ship-postal-code', 'ship-country'])

# Check for missing values after deleting na values
missing_values_1 = df_cleaned_rm_rows.isnull().sum()

# Display columns with their corresponding count of missing values
print(missing_values_1)
```

```python
# Replace the rest of the missing values to unkown
df_cleaned_rm_rows.loc[:,'Courier Status'] = df_cleaned_rm_rows['Courier Status'].fillna('Unknown')
df_cleaned_rm_rows.loc[:,'promotion-ids'] = df_cleaned_rm_rows['promotion-ids'].fillna('No Promotion Used')
df_cleaned_rm_rows.loc[:,'fulfilled-by'] = df_cleaned_rm_rows['fulfilled-by'].fillna('Unknown')
```

```python
# Now check for the missing values
missing_values_2 = df_cleaned_rm_rows.isnull().sum()

# Display columns with their corresponding count of missing values
print(missing_values_2)
```

```python
# We eliminated duplicate and missing values and before using it in SQL lets check if the data types are correct
print(df_cleaned_rm_rows.dtypes)
```

```python
# Convert 'Date' column to datetime
df_cleaned_rm_rows['Date'] = pd.to_datetime(df_cleaned_rm_rows['Date'], errors = 'coerce')

print(df_cleaned_rm_rows.dtypes)
```

```python
# Extracting Year and month from the data
df_cleaned_rm_rows.loc[:, 'Year'] = df_cleaned_rm_rows['Date'].dt.year
df_cleaned_rm_rows.loc[:, 'Month'] = df_cleaned_rm_rows['Date'].dt.month
```

```python
# Now let's see our final cleaned data
df_cleaned_rm_rows
```

```python
output_path = r'D:\Learning\Project\E-Commerce Project\Cleaned Amazon Sale Report.csv'
if os.path.exists(output_path):
    print("Output csv file already exists")
else:
    df_cleaned_rm_rows.to_csv(r'D:\Learning\Project\E-Commerce Project\Cleaned Amazon Sale Report.csv', index = False)
```
